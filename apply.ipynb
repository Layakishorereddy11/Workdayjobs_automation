```python
def login(driver,email1,password1):
    # Login code
    email = driver.find_element(By.XPATH, '//input[@data-automation-id="email"]')
    password = driver.find_element(By.XPATH, '//input[@data-automation-id="password"]')
    email.send_keys(email1)
    password.send_keys(password1)
    button = driver.find_element(By.XPATH, '//div[@aria-label="Sign In"]')
    button.click()

    # Wait and check for error/success
    wait = WebDriverWait(driver, 1.5)  # 10 seconds timeout
    try:
        # Wait for either error message or successful page load
        error_element = wait.until(
            EC.presence_of_element_located((By.XPATH, '//div[@data-automation-id="errorMessage"]'))
        )
        print("Invalid email or password")
        Createbutton = driver.find_element(By.XPATH, '//button[@data-automation-id="createAccountLink"]')
        Createbutton.click()
        Createemail = wait.until(
            EC.presence_of_element_located((By.XPATH, '//input[@data-automation-id="email"]'))
        )
        Createpassword = driver.find_element(By.XPATH, '//input[@data-automation-id="password"]')
        Verifypassword = driver.find_element(By.XPATH, '//input[@data-automation-id="verifyPassword"]')
        Createemail.send_keys(email1)
        Createpassword.send_keys(password1)
        Verifypassword.send_keys(password1)
        try:
        # Wait for checkboxes to be visible
            wait = WebDriverWait(driver, 10)
            checkboxes = wait.until(
                EC.presence_of_all_elements_located((By.XPATH, '//input[@type="checkbox"]'))
            )
            
            # Click all checkboxes that aren't checked
            for checkbox in checkboxes:
                if not checkbox.is_selected():
                    # Scroll to checkbox
                    driver.execute_script("arguments[0].scrollIntoView(true);", checkbox)
                    # Wait briefly for any animations
                    time.sleep(0.5)
                    # Click the checkbox
                    checkbox.click()
        except Exception as e:
            pass
        time.sleep(1)
        CreateAccountbutton = driver.find_element(By.XPATH, '//div[@aria-label="Create Account"]')
        CreateAccountbutton.click()
        
    except TimeoutException:
        # If error message doesn't appear within timeout, login was successful
        print("Login successful")
        
        

```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time

def scroll_to_element_advanced(driver, element):
    try:
        # Approach 1: Smooth scroll with behavior
        driver.execute_script(
            "arguments[0].scrollIntoView({behavior: 'smooth', block: 'center', inline: 'nearest'});", 
            element
        )

        
        # Approach 2: Action chains
        actions = ActionChains(driver)
        actions.move_to_element(element).perform()

        
        # Approach 3: Precise positioning with offset
        desired_y = (element.size['height'] / 2) + element.location['y']
        window_h = driver.execute_script('return window.innerHeight')
        window_y = driver.execute_script('return window.pageYOffset')
        current_y = (window_h / 2) + window_y
        scroll_y_by = desired_y - current_y
        
        driver.execute_script(f"window.scrollBy(0, arguments[0]);", scroll_y_by)

        
    except Exception as e:
        print(f"Scroll error: {str(e)}")
        # Fallback to simple scroll
        driver.execute_script("arguments[0].scrollIntoView(true);", element)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Initialize WebDriver (Make sure to have the correct WebDriver installed)
# driver = webdriver.Chrome()

# # Open a webpage
# driver.get("https://example.com")

# Find all input elements
input_elements = driver.find_elements(By.TAG_NAME, "input")

# Generate XPaths dynamically
def get_xpath(element):
    return driver.execute_script(
        """
        function getElementXPath(elt) {
            var path = [];
            for (; elt && elt.nodeType == 1; elt = elt.parentNode) {
                var idx = 1;
                for (var sib = elt.previousSibling; sib; sib = sib.previousSibling) {
                    if (sib.nodeType == 1 && sib.tagName == elt.tagName) idx++;
                }
                var xname = elt.tagName.toLowerCase();
                xname += "[" + idx + "]";
                path.unshift(xname);
            }
            return "/" + path.join("/");
        }
        return getElementXPath(arguments[0]);
        """,
        element,
    )

# Get associated question (label or placeholder) and last preceding span text
def get_associated_question(element):
    label = None
    try:
        # Check for associated label using 'for' attribute
        label_element = driver.find_element(By.XPATH, f"//label[@for='{element.get_attribute('id')}']")
        label = label_element.text if label_element else None
    except:
        pass
    
    if not label:
        # Check placeholder attribute as a fallback
        label = element.get_attribute("placeholder")
    
    return label if label else "Unknown"

# Get the last preceding span text before the input field
def get_last_preceding_span_text(element):
    span_text = None
    try:
        span_elements = element.find_elements(By.XPATH, "preceding::span")
        if span_elements:
            span_text = span_elements[-1].text  # Get the last span text
    except:
        pass
    
    return span_text if span_text else "No Span Text"

# Get input type
def get_input_type(element):
    input_type = element.get_attribute("type")
    return input_type if input_type else "Unknown"

# Check if input is untouched
def is_untouched(element):
    value = element.get_attribute("value")
    return value is None or value.strip() == ""

# Store XPaths, associated questions, last preceding span texts, and input types for untouched inputs
input_data = []
for input_element in input_elements:
    if is_untouched(input_element):  # Only consider untouched inputs
        xpath = get_xpath(input_element)
        question = get_associated_question(input_element)
        span_text = get_last_preceding_span_text(input_element)
        input_type = get_input_type(input_element)
        input_data.append((span_text, question, input_type, xpath))

# Print results for untouched inputs
for span_text, question, input_type, xpath in input_data:
    print(f"Last Span Text: {span_text}, Question: {question}, Input Type: {input_type}, XPath: {xpath}")

# # Close WebDriver
# driver.quit()
```
